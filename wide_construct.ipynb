```python
import pandas as pd
```

```python
LEAs = set()
schools = set()
rev_funds = set()
expense_funds = set()
```

```python
rev_filename = "revenue/revenue.csv"
df_rev = pd.read_csv(rev_filename)
```

```python
df_rev.rename(columns={'District Number': 'LEA_num', 'District Name': 'LEA', 'Location School Id':
                           'school_id', 'Location School Number': 'school_num', 'Location Name': 'school',
                           'Fund Code': 'fund_code', 'Revenue Code': 'rev_code', 'Program Code': 'program_code',
                           'Amount': 'rev_amt', 'Year': 'year'}, inplace=True)
df_rev.dropna(inplace=True)

df_rev['fund_code'] = df_rev['fund_code'].astype('int').astype('str')
df_rev['rev_code'] = df_rev['rev_code'].astype('int').astype('str')
df_rev['program_code'] = df_rev['program_code'].astype('int').astype('str')
```

```python
# create a list of all funds
print(df_rev['rev_code'].nunique())
rev_funds.update(set(df_rev['rev_code'].tolist()))
print(rev_funds)
LEAs.update(set(df_rev['LEA']))
```

```python
# wide format by revenue code
df_rev_temp = df_rev.copy()
df_rev_temp['rev_code'] = "r_" + df_rev_temp['rev_code']
df_rev_1 = df_rev_temp.groupby(['LEA', 'school', 'year', 'rev_code'])['rev_amt'].sum().unstack('rev_code')
df_rev_1.to_csv('rev_group_by_1.csv')

# wide format by revenue code and fund code
df_rev['fund_id'] = "r_" + df_rev['fund_code'] + "_" + df_rev['rev_code']
df_rev_2 = df_rev.groupby(['LEA', 'school', 'year', 'fund_id'])['rev_amt'].sum().unstack('fund_id')
df_rev_2.to_csv('rev_group_by_2.csv')

# wide format by all subcategories (big!)
df_rev['fund_id'] = "r_" + df_rev['fund_code'] + "_" + df_rev['rev_code'] + "_" + df_rev['program_code']
df_rev_3 = df_rev.groupby(['LEA', 'school', 'year', 'fund_id'])['rev_amt'].sum().unstack('fund_id')
df_rev_3.to_csv("rev_group_by_3.csv")
```

```python
exp_filename = "expense/expense.csv"
df_exp = pd.read_csv(exp_filename)
```

```python
df_exp.rename(columns={'District Number': 'LEA_num', 'District Name': 'LEA', 'Location School Id':
    'school_id', 'Location School Number': 'school_num', 'Location Name': 'school',
                       'Fund Code': 'fund_code', 'Function Code': 'exp_code', 'Program Code': 'program_code',
                       'Amount': 'exp_amt', 'Year': 'year'}, inplace=True)

df_exp.dropna(inplace=True)

df_exp['fund_code'] = df_exp['fund_code'].astype('int').astype('str')
df_exp['exp_code'] = df_exp['exp_code'].astype('int').astype('str')
df_exp['program_code'] = df_exp['program_code'].astype('int').astype('str')
```

```python
# create a list of all funds
print(df_exp['exp_code'].nunique())
expense_funds.update(set(df_exp['exp_code'].tolist()))
print(expense_funds)
LEAs.update(set(df_exp['LEA']))
```

```python
# wide format by revenue code
df_exp_temp = df_exp.copy()
df_exp_temp['exp_code'] = "e_" + df_exp_temp['exp_code']
df_exp_1 = df_exp_temp.groupby(['LEA', 'school', 'year', 'exp_code'])['exp_amt'].sum().unstack('exp_code')
df_exp_1.to_csv("exp_group_by_1.csv")

# wide format by revenue code and fund code
df_exp['fund_id'] = "e_" + df_exp['fund_code'] + "_" + df_exp['exp_code']
df_exp_2 = df_exp.groupby(['LEA', 'school', 'year', 'fund_id'])['exp_amt'].sum().unstack('fund_id')
df_exp_2.to_csv("exp_group_by_2.csv")

# wide format by all subcategories (big!)
df_exp['fund_id'] = "e_" + df_exp['fund_code'] + "_" + df_exp['exp_code'] + "_" + df_exp['program_code']
df_exp_3 = df_exp.groupby(['LEA', 'school', 'year', 'fund_id'])['exp_amt'].sum().unstack('fund_id')
df_exp_3.to_csv("exp_group_by_3.csv")
```

```python
df_merge_1 = df_rev_1.merge(df_exp_1, how='left', on=['LEA', 'school', 'year'])
df_merge_1.to_csv('merge_1.csv')

df_merge_2 = df_rev_2.merge(df_exp_2, how='left', on=['LEA', 'school', 'year'])
df_merge_2.to_csv('merge_2.csv')

df_merge_3 = df_rev_3.merge(df_exp_3, how='left', on=['LEA', 'school', 'year'])
df_merge_3.to_csv('merge_3.csv')
```

```python
merges = [df_merge_1, df_merge_2, df_merge_3]

for d in merges:
    num_rows = len(d.index)
    for col in d.columns:
        print(col + " " + str(num_rows - d[col].isna().sum()))
```
